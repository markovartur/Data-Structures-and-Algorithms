```python
#Бинарный поиск
from random import randint

start_time = time.time()

# Поиск
def binary_search_index(array, key):
    n = len(array)
    lower_bound, upper_bound = 0, n - 1
    middle = (lower_bound + upper_bound) // 2
    while upper_bound - lower_bound > 1:
        if key <= array[middle]:
            upper_bound = middle
        else:
            lower_bound = middle
        middle = (lower_bound + upper_bound) // 2
    return upper_bound

def binary_search_find(array, key):
    k = binary_search_index(array, key)
    assert key == array[k], "Искомый элемент не найден"
    return k

# Вставка
def binary_search_insert(array, key):
    array.insert(binary_search_index(array, key), key)
    return array

# Удаление
def binary_search_remove(array, key):
    element = binary_search_find(array, key)
    array.pop(element)
    return array


print('Исходный массив')
print(array)

print('Отсортированный массив')
print(sorted_array)

print('Вставка элемента 50 в массив')
print(binary_search_insert(sorted_array, 50))

print('Поиск элемента 50 в массиве')
print(binary_search_find(sorted_array, 50))

print('Удаление элемента 50 из массива')
print(binary_search_remove(sorted_array, 50))

time_binary_search = round((time.time() - start_time)*1000)
print("Время выполнения поиска")
print("--- {:.3f} ms ---".format(time_binary_search))
```

```python
#Метод Фибоначчи
start_time = time.time()

def Fibonaccian_Search(arr, x, n):
 
    fibMMm2 = 0  
    fibMMm1 = 1  
    fibM = fibMMm2 + fibMMm1  
 
    while (fibM < n):
        fibMMm2 = fibMMm1
        fibMMm1 = fibM
        fibM = fibMMm2 + fibMMm1
 
    offset = -1
 
    while (fibM > 1):
 
        i = min(offset+fibMMm2, n-1)
 
        if (arr[i] < x):
            fibM = fibMMm1
            fibMMm1 = fibMMm2
            fibMMm2 = fibM - fibMMm1
            offset = i
 
        elif (arr[i] > x):
            fibM = fibMMm2
            fibMMm1 = fibMMm1 - fibMMm2
            fibMMm2 = fibM - fibMMm1
 
        else:
            return i
 
    if(fibMMm1 and arr[n-1] == x):
        return n-1
 
    return -1
 
arr = [10, 22, 35, 40, 45, 50,
       80, 82, 85, 90, 100,235]
n = len(arr)
x = 235
ind = Fibonaccian_Search(arr, x, n)
if ind>=0:
  print("Found at index:",ind)
else:
  print(x,"isn't present in the array");

time_Fibonaccian_Search = round((time.time() - start_time)*1000)
print("--- {:.3f} ms ---".format(time_Fibonaccian_Search))
```

```python
#Интерполяционный
start_time = time.time()

def interpolation_search_index(lys, val):
    low = 0
    high = (len(lys) - 1)
    while low <= high and val >= lys[low] and val <= lys[high]:
        index = low + int(((float(high - low) / ( lys[high] - lys[low])) * ( val - lys[low])))
        if lys[index] == val:
            return index
        if lys[index] < val:
            low = index + 1;
            guess = low
        else:
            high = index - 1;
            guess = high+1
    return guess

def interpolation_search_find(lst, search_item):
    k = interpolation_search_index(lst, search_item)
    assert search_item == lst[k], "Искомый элемент не найден"
    return k

time_interpolation_search = round((time.time() - start_time)*1000)

# Вставка
def interpolation_search_insert(lst, search_item):
    lst.insert(interpolation_search_index(lst, search_item), search_item)
    return lst

# Удаление
def interpolation_search_remove(lst, search_item):
    element = interpolation_search_find(lst, search_item)
    lst.pop(element)
    return lst


print('Исходный массив')
print(array)

print('Отсортированный массив')
print(sorted_array)

print('Вставка элемента 50 в массив')
print(interpolation_search_insert(sorted_array, 50))

print('Поиск элемента 50 в массиве')
print(interpolation_search_find(sorted_array, 50))

print('Удаление элемента 50 из массива')
print(interpolation_search_remove(sorted_array, 50))

time_interpolation_search = round((time.time() - start_time)*1000)
print("Время выполнения поиска")
print("--- {:.3f} ms ---".format(time_interpolation_search))
```
